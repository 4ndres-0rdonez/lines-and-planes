```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
%matplotlib widget
```

```python
def create_line(x0, y0, z0, vx, vy, vz, t_min, t_max, t_N):
    # Parameter
    t = np.linspace(t_min, t_max, t_N)
    
    # Parametric equations of the line
    x = x0 + vx * t
    y = y0 + vy * t
    z = z0 + vz * t

    return x,y,z

def plot_line(x,y,z, label, color, ax=None):
    if ax==None:
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_zlabel('z')
        
    ax.plot(x, y, z, label=label, color=color)
    ax.legend()
        
    return ax      

def plot_point(x,y,z, label, color, ax=None):
    if ax==None:
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_zlabel('z')
        
    ax.scatter(x, y, z, color=color)    
    ax.text(x, y, z, '{}({},{},{})'.format(label,x,y,z))
    
    return ax      
```

```python
P1x, P1y, P1z = 2,1,-5
v1x, v1y, v1z = -1,5,-4
t1_min, t1_max, t1_N = -2, 1, 10
label1, color1 = 'line 1', 'green'
X1,Y1,Z1 = create_line(P1x,P1y,P1z, v1x,v1y,v1z, t1_min,t1_max,t1_N)
```

```python
P2x, P2y, P2z = -2,3,1
v2x, v2y, v2z = -2,2,-4
t2_min, t2_max, t2_N = -2, 1, 10
label2, color2 = 'line 2', 'blue'
X2,Y2,Z2 = create_line(P2x,P2y,P2z, v2x,v2y,v2z, t2_min,t2_max,t2_N)
```

```python
ax = plot_line(X1,Y1,Z1, label1, color1)
ax = plot_line(X2,Y2,Z2, label2, color2, ax=ax)
ax = plot_point(P1x, P1y, P1z, label=r'P1', color=color1, ax=ax)
ax = plot_point(P2x, P2y, P2z, label='P2', color=color2, ax=ax)

Q1x, Q1y, Q1z = 3,-4,-1
Q2x, Q2y, Q2z = 0, 1, 5
ax = plot_point(Q1x, Q1y, Q1z, label='Q1', color=color1, ax=ax)
ax = plot_point(Q2x, Q2y, Q2z, label='Q2', color=color2, ax=ax)
```

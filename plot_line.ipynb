```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
%matplotlib widget
```

```python
# Define the parameters for the line
x0, y0, z0 = 2, 1, 3  # Starting point
a, b, c = 1, 2, 3      # Direction of the line

# Define the parameter t
t = np.linspace(-2, 2, 50)  # Values of t from -5 to 5

# Parametric equations of the line
x = x0 + a * t
y = y0 + b * t
z = z0 + c * t
```

```python
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the line
ax.plot(x, y, z, label='3D Line', color='b')
ax.scatter(x0, y0, z0, color='r')  # Mark the starting point
ax.text(x0, y0, z0, '({},{},{})'.format(x0,y0,z0), color='red')

# Plot the origin
ax.scatter(0, 0, 0, color='k')  # Mark the starting point
ax.text(0, 0, 0, ' (0,0,0)', color='black')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show legend
ax.legend()

# Show the plot
plt.show()
```
